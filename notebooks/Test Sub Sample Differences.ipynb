#### Q1. Is there a significant difference in the average value_sales between two different abc_class_code (Top Sellers vs Low Sellers) categories?

To determine if there is a significant difference in the average value_sales between Top Sellers and Low Sellers (classified by abc_class_code as 'A' for Top Sellers and 'C' for Low Sellers), I will perform a two-sample t-test to compare the means of these two independent groups.

**Formulate Hypotheses:**

Null Hypothesis (H₀) -  The mean value_sales is the same for Top Sellers and Low Sellers. (H₀:μTopSellers =μLowSellers)

Alternative Hypothesis (H₁) - The mean value_sales is different between Top Sellers and Low Sellers. (H₁:μTopSellers =/ μLowSellers)

​**Analysis Approach** If the p-value from the t-test is less than the chosen significance level (usually 0.05), we will reject the null hypothesis, suggesting a significant difference in the average value_sales between Top Sellers and Low Sellers. If the p-value is greater than 0.05, we will fail to reject the null hypothesis, indicating no significant difference in the means of these two groups.

```python

abc_class = sample_test[sample_test['abc_class_code'].isin(['A', 'C'])]

# # Plotting histogram for visualization
# sns.histplot(data=abc_class, x='value_sales', bins=50, kde=True, log_scale=True, hue='abc_class_code')
# plt.title("Histogram of value_sales (Log scale)")
# plt.legend(title='abc class code', loc='upper left', labels=['Top Sellers', 'Low Sellers'])
# plt.show()

# Separating data into two groups for the t-test
top_sellers_sales = abc_class[abc_class['abc_class_code'] == 'A']['value_sales']
low_sellers_sales = abc_class[abc_class['abc_class_code'] == 'C']['value_sales']

# Performing the two-sample t-test
t_stat, p_value = stats.ttest_ind(top_sellers_sales, low_sellers_sales, equal_var=False)  # Using Welch's t-test for unequal variance

# Displaying the test results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("The mean value_sale are significantly different between Topsaller and lowseller.")
else:
    print("There is no significant difference in value_sale between Topsaller and lowseller.")
```
