```python
# Filter only numeric and boolean columns for correlation calculation
data_numeric = data.select_dtypes(include=['number', 'bool'])

# Calculate the correlation matrix
corr_matrix = data_numeric.corr()

# Show correlations with 'value_sales', sorted by absolute value in descending order
top_10_corr_with_sales = corr_matrix['value_sales'].abs().sort_values(ascending=False).head(10)
print(top_10_corr_with_sales)
```
